**LABELLING CATEGORIES IN NUMERICAL FEATURES**

```python
df_revised["Education"] = df_revised["Education"].replace({1:"Below College",2:"College",3:"Bachelor",4:"Master",5:"Doctor"})
df_revised["EnvironmentSatisfaction"] = df_revised["EnvironmentSatisfaction"].replace({1:"Low",2:"Medium",3:"High",4:"Very High"})
df_revised["JobInvolvement"] = df_revised["JobInvolvement"].replace({1:"Low",2:"Medium",3:"High",4:"Very High"})
df_revised["JobLevel"] = df_revised["JobLevel"].replace({1:"Entry Level",2:"Junior Level",3:"Mid Level",4:"Senior Level",5:"Executive Level"})
df_revised["JobSatisfaction"] = df_revised["JobSatisfaction"].replace({1:"Low",2:"Medium",3:"High",4:"Very High"})
df_revised["PerformanceRating"] = df_revised["PerformanceRating"].replace({1:"Low",2:"Good",3:"Excellent",4:"Outstanding"})
df_revised["RelationshipSatisfaction"] = df_revised["RelationshipSatisfaction"].replace({1:"Low",2:"Medium",3:"High",4:"Very High"})
df_revised["WorkLifeBalance"] = df_revised["WorkLifeBalance"].replace({1:"Bad",2:"Good",3:"Better",4:"Best"})
```

ENLISTING ONLY CATEGORICAL FEATURES

**DROPPING UNNECESSARY COLUMNS**

```python
df_revised.drop(['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'], axis="columns", inplace=True)
```
